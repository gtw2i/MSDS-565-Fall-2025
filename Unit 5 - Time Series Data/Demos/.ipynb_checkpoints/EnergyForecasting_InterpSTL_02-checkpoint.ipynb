```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)
pd.set_option("display.precision", 2)

plt.style.use('dark_background')
```

```python
df = pd.read_csv('../../data/Hourly Energy Consumption/DAYTON_hourly.csv',
                 index_col=0,
                 header=0)

df.index = pd.to_datetime(df.index)

df = df.rename(columns={'DAYTON_MW':'mw'})

dates = pd.Series(pd.date_range(start=df.index.min(),
                                end=df.index.max(),
                                freq='H'),
                  name='date')

df = pd.merge(df, dates, left_index=True, right_on='date', how='left')
df = df.set_index('date').sort_index()
df = df.ffill()
```

```python
df.resample('w').sum().plot(figsize=(10,4))
```

```python
df1 = df.resample('d').sum()
df1 = df1.iloc[1:-1].copy()

df1 = df1[ df1.index.year >= 2009 ]

df1.plot(figsize=(10,4))
```

```python
from statsmodels.tsa.seasonal import STL

y_STL = df1.mw.copy()

# Perform STL decomposition
stl = STL(y_STL, period=7)
result = stl.fit()

y_trend    = result.trend
y_seasonal = result.seasonal
y_resid    = result.resid
```

```python
plt_idx1 = 7*52

plt.figure(figsize=(15, 12))

plt.subplot(4, 1, 1)
plt.plot(y_STL.iloc[:plt_idx1], 'r', label='Original Data')
plt.plot(y_trend.iloc[:plt_idx1], 'g', label='Trend')
#plt.ylim(df1.mw.min(),df1.mw.max())
plt.legend()
plt.title('Original Time Series Data')

plt.subplot(4, 1, 2)
plt.plot(y_trend.iloc[:plt_idx1], label='Trend')
plt.legend()
plt.title('Trend Component')

plt.subplot(4, 1, 3)
plt.plot(y_seasonal.iloc[:plt_idx1], label='Seasonal')
#plt.ylim(df1.mw.min(),df1.mw.max())
plt.legend()
plt.title('Seasonal Component')

plt.subplot(4, 1, 4)
plt.plot(y_resid.iloc[:plt_idx1], label='Residual')
#plt.ylim(df1.mw.min(),df1.mw.max())
plt.legend()
plt.title('Residual Component')

plt.tight_layout()
plt.show()
```

```python
df1['mw_t']  = y_trend
df1['mw_ts'] = y_trend + y_seasonal

df1
```

```python
df1[:7*52].plot(figsize=(14,6),
                color=['red','green','blue'])
```

```python
df2 = pd.DataFrame()

col = 'mw_t'

df2['mw_raw'] = df1[col].resample('h').mean()
df2['mw_lin'] = df1[col].resample('h').mean().interpolate(method='linear')
df2['mw_cub'] = df1[col].resample('h').mean().interpolate(method='cubic')
df2['mw_ak']  = df1[col].resample('h').mean().interpolate(method='akima')
df2['mw_pc']  = df1[col].resample('h').mean().interpolate(method='pchip')


df2.head(30)
```

```python
plt.figure(figsize=(12,4))

plt_idx2 = 24*7*52

#col = 'mw_lin'
col = 'mw_cub'

sns.lineplot( x=df2.iloc[:plt_idx2].mw_raw.index, y=df2.iloc[:plt_idx2][col], color='b', linewidth=1, label=col )
sns.scatterplot( x=df2.iloc[:plt_idx2].mw_raw.index, y=df2.iloc[:plt_idx2].mw_raw, color='r', linewidth=0, s=50, label='mw_raw' )

plt.legend()
```

# make features

```python
df2['hour']       = df2.index.hour
df2['dayofweek']  = df2.index.dayofweek
df2['dayofmonth'] = df2.index.day
df2['week']       = df2.index.isocalendar().week.astype(int)
df2['month']      = df2.index.month
df2['quarter']    = df2.index.quarter
df2['year']       = df2.index.year

df2
```

```python
from numpy import pi

for i in range(1,24//2):
    df2[f'hour_sin{i}'] = np.sin(i*2*pi*df2.hour/24)
    df2[f'hour_cos{i}'] = np.cos(i*2*pi*df2.hour/24)
# end

for i in range(1,7//2):
    df2[f'dayofweek_sin{i}'] = np.sin(i*2*pi*df2.dayofweek/7)
    df2[f'dayofweek_cos{i}'] = np.cos(i*2*pi*df2.dayofweek/7)
# end

"""
for i in range(1,31//2):
    df2[f'dayofmonth_sin{i}'] = np.sin(i*2*pi*df2.dayofmonth/31)
    df2[f'dayofmonth_cos{i}'] = np.cos(i*2*pi*df2.dayofmonth/31)
# end

for i in range(1,53//2):
    df2[f'year_sin{i}'] = np.sin(i*2*pi*df2.year/53)
    df2[f'year_cos{i}'] = np.cos(i*2*pi*df2.year/53)
# end
""";

for i in range(1,12//2+1):
    df2[f'month_sin{i}'] = np.sin(i*2*pi*df2.month/12)
    df2[f'month_cos{i}'] = np.cos(i*2*pi*df2.month/12)
# end

for i in range(1,4//2+1):
    df2[f'quarter_sin{i}'] = np.sin(i*2*pi*df2.quarter/4)
    df2[f'quarter_cos{i}'] = np.cos(i*2*pi*df2.quarter/4)
# end
```

```python
df2
```

```python
X = df2.iloc[:,5:]
y = df2['mw_cub']
```

```python
idx_split = 24*7*52*1
print(idx_split)

X_train = X.iloc[:-idx_split,:]
y_train = y[:-idx_split]

X_test  = X.iloc[-idx_split:,:]
y_test  = y[-idx_split:]

plt.figure(figsize=(7,3))
plt.plot( y_train, 'b' );
plt.plot( y_test, 'r' );
```

# model

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor

models = [RandomForestRegressor(n_estimators=100, max_depth=10),
          GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=2),
          GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3),
          GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=5),
         ]
```

```python
from sklearn.metrics import r2_score, mean_squared_error

for m in models:
    m.fit(X_train, y_train)
    
    y_train_pred = m.predict(X_train)
    y_test_pred  = m.predict(X_test)
    
    print(m)
    
    r2_train = r2_score(y_train, y_train_pred)
    print(f'r2_train:   {r2_train:.5}')

    r2_test = r2_score(y_test, y_test_pred)
    print(f'r2_test:    {r2_test:.5}')
    
    rmse_train = mean_squared_error(y_train, y_train_pred)**0.5
    print(f'rmse_train: {rmse_train:.5}')

    rmse_test = mean_squared_error(y_test, y_test_pred)**0.5
    print(f'rmse_test:  {rmse_test:.5}')
    
    mape_train = np.abs( (y_train - y_train_pred)/y_train ).mean()
    print(f'mape_train: {mape_train:.5}')

    mape_test = np.abs( (y_test - y_test_pred)/y_test ).mean()
    print(f'mape_test:  {mape_test:.5}')
    
    y_train_pred = pd.Series( y_train_pred, index=y_train.index )
    
    plt.figure(figsize=(12,5))
    plt.plot( y_train, 'r' );
    plt.plot( y_train_pred, 'g' );
    plt.show()
    
    y_test_pred = pd.Series( y_test_pred, index=y_test.index )

    plt.figure(figsize=(7,3))
    plt.plot( y_test, 'r' );
    plt.plot( y_test_pred, 'g' );
    plt.show()
    
    print()
# end
```
