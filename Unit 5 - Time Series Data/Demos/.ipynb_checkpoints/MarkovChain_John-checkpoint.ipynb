```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)
pd.set_option("display.precision", 2)

plt.style.use('dark_background')
```

# Read file

```python
with open('John.txt','r', encoding='utf-8') as f:
    text = f.read()
# end

print(text[:100])
```

# Clean text

```python
text = text.lower()
text = re.sub('\[\d+\]','',text)
text = re.sub('\s+',' ',text)
text = re.sub('[,;:\(\)\.\!\?]','',text)
text = re.sub('\.','',text)
text = re.sub('john\d+','',text)
text = text.strip()

print(text[:100])
```

# Encode text as integers

```python
words  = np.array(text.split(' '))
tokens = np.unique(words)

n_words  = words.shape[0]
n_tokens = tokens.shape[0]

from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
tokens_enc = LE.fit_transform(tokens).astype(int)
words_enc = LE.transform(words).astype(int)

df = pd.DataFrame({'words':words,'words_enc':words_enc})
df.value_counts()
```

# Get transition probabilities (kernel)

```python
P = np.zeros((n_tokens,n_tokens))

for i in range(n_words-1):
    jj = words_enc[i]
    kk = words_enc[i+1]
    P[jj,kk] += 1
# end
```

```python
for i in range(n_tokens):
    tot = P[i,:].sum()
    if tot > 0:
        P[i,:] = P[i,:] / tot
    # end
# end
```

```python
P[0,:].sum()
```

# Generate new text

```python
nStep = 25

start = 'the'
start_enc = LE.transform([start])[0]

chain_enc = [start_enc]
for i in range(nStep-1):
    curr = chain_enc[i]
    new = np.random.choice( tokens_enc, p=P[curr,:], size=1 )[0]
    chain_enc.append(new)
# end
chain_enc = np.array(chain_enc)

chain = LE.inverse_transform(chain_enc)

text_gen = ' '.join(chain)
print(text_gen)
```
