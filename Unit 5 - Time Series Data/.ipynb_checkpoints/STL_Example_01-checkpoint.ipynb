```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import STL

# Generate a sample time series data with datetime index
np.random.seed(42)
date_rng = pd.date_range(start='2022-01-01', end='2022-01-05', freq='D')
#data = 2 * np.sin(0.2 * np.arange(len(date_rng))) + np.arange(len(date_rng)) + np.random.normal(scale=0.1, size=len(date_rng))
data = 30*np.sin(0.2*np.arange(len(date_rng))) + np.arange(len(date_rng)) + np.random.normal(scale=0.1, size=len(date_rng))

# Create a pandas DataFrame with the time series data and datetime index
df = pd.DataFrame(data, columns=['value'], index=date_rng)

df = df.drop(df.index[2])
df
```

```python
x = pd.Series(pd.date_range(start=df.index.min(), end=df.index.max(), freq='D'), name='date')

df = pd.merge(df, x, left_index=True, right_on='date', how='outer')
df = df.set_index('date').sort_index()
df = df.fillna(df.median())
df
```

```python
# Perform STL decomposition
stl = STL(df['value'], seasonal=13)  # You can adjust the seasonal parameter
result = stl.fit()

# Plot the original time series and its components
plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 1)
plt.plot(df['value'], label='Original Data')
plt.legend()
plt.title('Original Time Series Data')

plt.subplot(4, 1, 2)
plt.plot(result.trend, label='Trend')
plt.legend()
plt.title('Trend Component')

plt.subplot(4, 1, 3)
plt.plot(result.seasonal, label='Seasonal')
plt.legend()
plt.title('Seasonal Component')

plt.subplot(4, 1, 4)
plt.plot(result.resid, label='Residual')
plt.legend()
plt.title('Residual Component')

plt.tight_layout()
plt.show()
```
