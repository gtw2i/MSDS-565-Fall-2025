# statsmodels linear reg

```python
import statsmodels.api as sm

n, _ = df.shape

y_train = y.iloc[:n//2]
X_train = X.iloc[:n//2]

X2 = sm.add_constant(X)
X2_train = X2.iloc[:n//2]

model = sm.OLS(y_train, X2_train)
results = model.fit()
pred = results.predict(X2)

results.summary()
```

```python
def plot(df, pred, n):
    df.DAYTON_MW.iloc[:n].plot(c='b', figsize=(15,5))
    pred.iloc[:n].plot(c='r')
# end

plot(df, pred, 24**365*5)
```

```python
# Apply the FFT to the time series data
fft_result = np.abs(np.fft.fft(s2))[1:len(s2)//2]

# Calculate the frequencies corresponding to the FFT result
# The result is in cycles per unit of your time series (cycles per hour in this case)
frequencies = np.fft.fftfreq(len(s2), 1)[1:len(s2)//2]

# Find the indices of the two largest amplitude peaks
indices_of_largest_peaks = np.argpartition(fft_result, -5)[-5:]

# Get the corresponding frequencies (periods)
periods_of_largest_peaks = 1 / frequencies[indices_of_largest_peaks]

# Plot the amplitude spectrum
plt.figure(figsize=(20, 10))
plt.plot(frequencies, fft_result)
plt.title('Frequency Domain Representation (FFT)')
plt.xlabel('Frequency (cycles per hour)')
plt.ylabel('Amplitude')

# Highlight the two largest peaks on the plot
plt.scatter(frequencies[indices_of_largest_peaks], fft_result[indices_of_largest_peaks], color='red', label='Largest Peaks')

plt.show()

# Print the periods of the two largest peaks
print("Periods of the two largest peaks:")
print(periods_of_largest_peaks)
```
