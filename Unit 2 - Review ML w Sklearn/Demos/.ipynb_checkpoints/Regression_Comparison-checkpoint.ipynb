```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score

plt.style.use('dark_background')
```

```python
nPts = 50

m = 0.5
b = 1

x = np.linspace(-2, 2, nPts)
X = x[:, np.newaxis]

#y = m*x + b + np.random.normal(0, 0.1, nPts)
y = np.sin(2*np.pi*x/2) + np.random.normal(0, 0.1, nPts)

plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Data')

plt.title('Regression Models Comparison')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
```

```python
# Linear Regression
linear_reg = LinearRegression(fit_intercept=True)
linear_reg.fit(X, y)
y_linear = linear_reg.predict(X)

# K-Nearest Neighbors Regression
knn_reg = KNeighborsRegressor(n_neighbors=5)
knn_reg.fit(X, y)
y_knn = knn_reg.predict(X)

# Decision Tree Regression
tree_reg = DecisionTreeRegressor()
tree_reg.fit(X, y)
y_tree = tree_reg.predict(X)

# Random Forest Regression
rf_reg = RandomForestRegressor(n_estimators=100)
rf_reg.fit(X, y)
y_rf = rf_reg.predict(X)

# Multi-layer Perceptron Regression
mlp_reg = MLPRegressor(hidden_layer_sizes=(100,), max_iter=1000)
mlp_reg.fit(X, y)
y_mlp = mlp_reg.predict(X)
```

```python
print(r2_score(y,y_linear))
print(r2_score(y,y_knn))
print(r2_score(y,y_tree))
print(r2_score(y,y_rf))
print(r2_score(y,y_mlp))
```

```python
plt.figure(figsize=(14, 7))
plt.scatter(x, y, label='Data with Noise')
plt.plot(x, y_linear, 'r', label='Linear Regression')
plt.plot(x, y_knn, 'g', label='KNN Regression')
plt.plot(x, y_tree, 'm', label='Decision Tree Regression')
plt.plot(x, y_rf, 'y', label='Random Forest Regression')
plt.plot(x, y_mlp, 'w', label='MLP Regression')

plt.title('Regression Models Comparison')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
```
