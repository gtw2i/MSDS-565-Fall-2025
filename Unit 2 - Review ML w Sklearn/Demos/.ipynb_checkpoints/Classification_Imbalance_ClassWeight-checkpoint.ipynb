```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)

plt.style.use('dark_background')
```

# Function for creating imbalanced data

```python
def GenerateData(n_pts1,n_pts2,center_sep,std):
    
    cov = std**2*np.eye(2)
    
    x1 = np.random.multivariate_normal(mean=[-center_sep/2,0],
                                       cov=cov,
                                       size=n_pts1)
    y1 = np.zeros(n_pts1)
    
    x2 = np.random.multivariate_normal(mean=[center_sep/2,0],
                                       cov=cov,
                                       size=n_pts2)
    y2 = np.ones(n_pts2)
    
    X = np.concatenate((x1,x2),axis=0)
    y = np.concatenate((y1,y2),axis=0)
    
    return X, y
# end
```

# Create data

```python
#np.random.seed(0)

n_pts0     = 1000
n_pts1     = 100
std        = 0.8
center_sep = 1.5

# generate data
X, y = GenerateData(n_pts0,
                    n_pts1,
                    center_sep,
                    std)

plt_scale = 3
plt.scatter(X[:,0], X[:,1], c=y, s=20, cmap='RdBu')
plt.xlim(-plt_scale,plt_scale)
plt.ylim(-plt_scale,plt_scale)
plt.gca().set_aspect('equal')
```

# Train/test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    stratify=y)
```

# Fit/predict/evaluate models

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

from sklearn.metrics import confusion_matrix, classification_report

class_weight = 'balanced'

models = [LogisticRegression(multi_class='ovr', class_weight=class_weight),
          DecisionTreeClassifier(max_leaf_nodes=10, class_weight=class_weight),
          RandomForestClassifier(n_estimators=100, max_leaf_nodes=10, class_weight=class_weight)]

for m in models:
    m.fit(X_train, y_train)
    
    y_pred = m.predict(X)
    y_train_pred = m.predict(X_train)
    y_test_pred  = m.predict(X_test)
    
    train_conf_matrix = confusion_matrix(y_train, y_train_pred)
    test_conf_matrix = confusion_matrix(y_test, y_test_pred)

    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(6,3));

    sns.heatmap(train_conf_matrix, annot=True, ax=axes[0], vmin=0, fmt='d');
    axes[0].set_title('Training');
    sns.heatmap(test_conf_matrix, annot=True, ax=axes[1], vmin=0, fmt='d');
    axes[1].set_title('Testing');
    fig.suptitle(str(m))
    plt.tight_layout()
    plt.show()
    
    print("Training")
    print(classification_report(y_train, y_train_pred))
    
    print("Testing")
    print(classification_report(y_test, y_test_pred))
    
# end
```
