```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)

plt.style.use('dark_background')
```

# Make data

```python
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=500,
                  centers=3,
                  cluster_std=0.4,
                  random_state=0)

X -= np.mean(X,axis=0)

cmap = 'turbo'
MAX = np.max(np.abs(X))*1.2
plt.scatter(X[:,0], X[:,1], c='w', s=5);
plt.xlim(-MAX,MAX)
plt.ylim(-MAX,MAX)
plt.gca().set_aspect('equal')
```

```python
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=500,
                  centers=3,
                  cluster_std=0.6,
                  random_state=1)

transformation = [[0.6, -0.6], [-0.4, 0.8]]
X = np.dot(X, transformation)

X -= np.mean(X,axis=0)

cmap = 'turbo'
MAX = np.max(np.abs(X))*1.2
plt.scatter(X[:,0], X[:,1], c='w', s=5);
plt.xlim(-MAX,MAX)
plt.ylim(-MAX,MAX)
plt.gca().set_aspect('equal')
```

```python
from sklearn.datasets import make_circles

X, _ = make_circles(n_samples=500,
                    factor=0.5,
                    noise=0.05)

X -= np.mean(X,axis=0)

cmap = 'turbo'
MAX = np.max(np.abs(X))*1.2
plt.scatter(X[:,0], X[:,1], c='w', s=5);
plt.xlim(-MAX,MAX)
plt.ylim(-MAX,MAX)
plt.gca().set_aspect('equal')
```

```python
from sklearn.datasets import make_circles

X = np.random.random((500,2))

X -= np.mean(X,axis=0)

cmap = 'turbo'
MAX = np.max(np.abs(X))*1.2
plt.scatter(X[:,0], X[:,1], c='w', s=5);
plt.xlim(-MAX,MAX)
plt.ylim(-MAX,MAX)
plt.gca().set_aspect('equal')
```

# Perform DBSCAN w/ sklearn

### like agglomerative, you can't generalize the model to new data

```python
from sklearn.cluster import DBSCAN

radius = 0.1

# Find clusters
model = DBSCAN(eps=radius, min_samples=5)
labels = model.fit_predict(X)

print(pd.Series(labels).value_counts().sort_index())

# Plot
plt.figure(figsize=(10,10))

theta = np.linspace(0,2*np.pi,100)
c = radius * np.cos(theta)
s = radius * np.sin(theta)

plt.plot(c, s, linewidth=1, c='w')

plt.scatter(X[:,0], X[:,1], c=labels, s=10, cmap=cmap);
plt.xlim(-MAX,MAX)
plt.ylim(-MAX,MAX)
plt.gca().set_aspect('equal')
```

# Varying the radius

```python
radii = 0.02 * (np.arange(25)+1)

n_clusters = []

for r in radii:
    model = DBSCAN(eps=r, min_samples=5)
    labels = model.fit_predict(X)
    labels = pd.Series(labels)
    
    vc = labels.value_counts()
    try:
        n_clusters.append( vc.shape[0] - 1 + vc[-1] )
    except:
        n_clusters.append( vc.shape[0] )
    # end
# end

plt.plot( radii, n_clusters )
```

# Finding the ideal epsilon radius

### manually

```python
from sklearn.neighbors import NearestNeighbors

n_neighbors = int(X.shape[0]*0.1)
#n_neighbors = 3
model_neigh = NearestNeighbors(n_neighbors=n_neighbors)
model_neigh.fit(X)

distances, indices = model_neigh.kneighbors(X,
                                            n_neighbors=n_neighbors,
                                            return_distance=True)

# Compute the median distance to the k-nearest neighbors
radius = np.median(distances[:, -1])
radius
```

### using estimate bandwidth

```python
from sklearn.cluster import estimate_bandwidth

radius = estimate_bandwidth(X, quantile=0.1)
radius
```

### test

```python
from sklearn.cluster import DBSCAN

# Find clusters
model = DBSCAN(eps=radius, min_samples=5)
labels = model.fit_predict(X)

print(pd.Series(labels).value_counts().sort_index())

# Plot
plt.figure(figsize=(10,10))

theta = np.linspace(0,2*np.pi,100)
c = radius * np.cos(theta)
s = radius * np.sin(theta)

plt.plot(c, s, linewidth=1, c='w')

plt.scatter(X[:,0], X[:,1], c=labels, s=10, cmap=cmap);
plt.xlim(-MAX,MAX)
plt.ylim(-MAX,MAX)
plt.gca().set_aspect('equal')
```
