```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)

plt.style.use('dark_background')
```

# Create data

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=1000,
                  n_features=2,
                  centers=2,
                  cluster_std=4.0,
                  random_state=6)

cmap = 'jet'

MAX = np.max(np.abs(X))

plt.scatter(X[:,0], X[:,1], c=y, s=10, cmap=cmap);
plt.xlim(-MAX,MAX)
plt.ylim(-MAX,MAX)
plt.gca().set_aspect('equal')
```

# Train/test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# Fit/predict model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

n_neighbors = 5
model = KNeighborsClassifier(n_neighbors=n_neighbors)

#max_leaf_nodes = 3
#model = DecisionTreeClassifier(max_leaf_nodes=max_leaf_nodes)

model.fit(X_train, y_train)

y_pred = model.predict(X)
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

print(f'Accuracy (train):   {train_accuracy}')
print(f'Accuracy (test):    {test_accuracy}')
print()

train_precision = precision_score(y_train, y_train_pred, average=None)
test_precision = precision_score(y_test, y_test_pred, average=None)

print(f'Training Precision: {train_precision}')
print(f'Testing Precision:  {test_precision}')
print()

train_recall = recall_score(y_train, y_train_pred, average=None)
test_recall = recall_score(y_test, y_test_pred, average=None)
print(f'Training Recall:    {train_recall}')
print(f'Testing Recall:     {test_recall}')
print()

train_f1 = f1_score(y_train, y_train_pred, average=None)
test_f1 = f1_score(y_test, y_test_pred, average=None)
print(f'Training F1 Score:  {train_f1}')
print(f'Testing F1 Score:   {test_f1}')
print()

train_conf_matrix = confusion_matrix(y_train, y_train_pred)
test_conf_matrix = confusion_matrix(y_test, y_test_pred)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8,3));

sns.heatmap(train_conf_matrix, annot=True, ax=axes[0]);
axes[0].set_title('Training');
sns.heatmap(test_conf_matrix, annot=True, ax=axes[1]);
axes[1].set_title('Testing');
```

```python
from sklearn.metrics import  classification_report

print(classification_report(y, y_pred))
```

```python
from sklearn.metrics import precision_recall_fscore_support

precision_recall_fscore_support(y, y_pred)
```

# ROC Curve (receiver operating characteristic) and AUC (area under curve)

```python
y_prob = model.predict_proba(X)
y_prob
```

```python
from sklearn.metrics import RocCurveDisplay

RocCurveDisplay.from_estimator(model, X_test, y_test)
#plt.gca().set_aspect('equal')
```

### AUC

```python
from sklearn.metrics import roc_curve, auc, roc_auc_score

fpr, tpr, thresholds = roc_curve(y, y_prob[:,1])
auc(fpr, tpr)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y, y_prob[:,1])
```

### Manual calculation

```python
y_test_prob = model.predict_proba(X_test)
#y_test_prob = np.random.rand(y_test.shape[0],3)
#y_test_prob = y_test_prob / y_test_prob.sum(axis=1).reshape(-1,1)

nPts = 10

thresh = np.linspace(0,1.001,nPts)

FPR = []
TPR = []
for t in thresh:
    y_test_roc = (y_test_prob[:,1] >= t).astype(int)
    cm = confusion_matrix(y_test, y_test_roc)
    #print(cm)
    
    tn = cm[0,0]
    fn = cm[1,0]
    fp = cm[0,1]
    tp = cm[1,1]
    
    fpr = fp/(fp+tn) if (fp+tn)>0 else 0
    tpr = tp/(tp+fn) if (tp+fn)>0 else 0
    
    FPR.append(fpr)
    TPR.append(tpr)
# end
FPR = np.array(FPR)
TPR = np.array(TPR)

plt.plot(FPR,TPR, '.');
plt.plot([0,1], [0,1], 'r--');
#plt.gca().set_aspect('equal')
```

### AUC (in accurate since I'm not using the trapezoidal rule for integration)

```python
np.sum(0.5*(TPR[:-1] + TPR[1:])*(FPR[:-1] - FPR[1:]))
```
