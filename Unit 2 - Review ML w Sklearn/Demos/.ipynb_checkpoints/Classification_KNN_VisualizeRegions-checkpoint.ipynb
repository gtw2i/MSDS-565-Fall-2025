```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)

plt.style.use('dark_background')
```

# Create data

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=100,
                  n_features=2,
                  centers=2,
                  cluster_std=20,
                  random_state=6)

X -= np.mean(X,axis=0)

cmap = 'jet'

MAX = np.max(np.abs(X))

plt.scatter(X[:,0], X[:,1], c=y, s=50, cmap=cmap);
plt.xlim(-MAX,MAX)
plt.ylim(-MAX,MAX)
plt.gca().set_aspect('equal')
```

```python
print(y)
```

# Train/test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,6))

axes[0].scatter(X_train[:,0], X_train[:,1], c=y_train, s=20, cmap=cmap);
axes[0].set_xlim(-MAX,MAX)
axes[0].set_ylim(-MAX,MAX)
axes[0].set_aspect('equal')
axes[0].set_title('Training data');

axes[1].scatter(X_test[:,0], X_test[:,1], c=y_test, s=20, cmap=cmap);
axes[1].set_xlim(-MAX,MAX)
axes[1].set_ylim(-MAX,MAX)
axes[1].set_aspect('equal')
axes[1].set_title('Testing data');
```

# Fit/predict model

```python
from sklearn.neighbors import KNeighborsClassifier

n_neighbors = 10
model = KNeighborsClassifier(n_neighbors=n_neighbors)

model.fit(X_train, y_train)

# get class predictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
y_test
```

```python
y_test_pred
```

```python
# get prediction probabilities
y_test_prob = model.predict_proba(X_test)
y_test_prob
```

```python
# get class predictions by finding the class with highest probability
y_test_prob.argmax(axis=1)
```

# Check accuracy

```python
from sklearn.metrics import accuracy_score, confusion_matrix

train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

print(f'Accuracy (train): {train_accuracy}')
print(f'Accuracy (test):  {test_accuracy}')
print()

train_conf_matrix = confusion_matrix(y_train, y_train_pred)
test_conf_matrix = confusion_matrix(y_test, y_test_pred)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8,3));

sns.heatmap(train_conf_matrix, annot=True, ax=axes[0]);
sns.heatmap(test_conf_matrix, annot=True, ax=axes[1]);
```

# Visualize prediction regions

```python
nGrid = 200
a1 = np.linspace(-MAX,MAX,nGrid)
a2 = np.linspace(-MAX,MAX,nGrid)

A1, A2 = np.meshgrid(a1,a2)
A1 = A1.flatten()
A2 = A2.flatten()
A = np.vstack((A1,A2)).T

B = model.predict(A)
B = B.reshape(nGrid,nGrid)
B = np.flipud(B)

extent = [-MAX,MAX,-MAX,MAX]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,6))

axes[0].imshow(B, interpolation='none', extent=extent,cmap=cmap, alpha=0.4)
axes[0].scatter(X_train[:,0], X_train[:,1], c=y_train, s=10, cmap=cmap);
axes[0].set_aspect('equal')
axes[0].set_title('Training data');

axes[1].imshow(B, interpolation='none', extent=extent,cmap=cmap, alpha=0.4)
axes[1].scatter(X_test[:,0], X_test[:,1], c=y_test, s=10, cmap=cmap);
axes[1].set_aspect('equal')
axes[1].set_title('Testing data');
```
