```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)

plt.style.use('dark_background')
```

```python
df = pd.read_csv('../../../CSCI 4900/data/Real Estate/Real Estate Data.csv', index_col=0, header=0)
df = df.reset_index(drop=True)

df = df.rename(columns = {'Type':'Dwelling Type',
                          'OvQual':'Overall Qual',
                          'Nbhd':'Neighborhood',
                          'Built':'Year Built'})

df.info()
df.head()
```

```python
gb1 = df.groupby(["Neighborhood"])['Lot Frontage'].median().sort_values()

def fill_1(r):
    n = r.Neighborhood
    
    x = r['Lot Frontage']
    
    if pd.isna(x):
        return gb1[n]
    else:
        return x
    # end
# end

df['Lot Frontage'] = df.apply(fill_1, axis=1)
```

```python
gb2 = df.groupby(["Neighborhood"])['Garage Yr Built'].median().sort_values()

def fill_2(r):
    n = r.Neighborhood
    
    x = r['Garage Yr Built']
    
    if pd.isna(x):
        return gb2[n]
    else:
        return x
    # end
# end

df['Garage Yr Built'] = df.apply(fill_2, axis=1)
```

```python
gb3 = df.groupby(["Neighborhood"])['Masonry/Veneer Area'].median().sort_values()

def fill_3(r):
    n = r.Neighborhood
    
    x = r['Masonry/Veneer Area']
    
    if pd.isna(x):
        return gb2[n]
    else:
        return x
    # end
# end

df['Masonry/Veneer Area'] = df.apply(fill_3, axis=1)
```

```python
df = df.select_dtypes(exclude=object)

X = df.drop(columns=['Sale Price'])
y = df['Sale Price']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)

X = pd.DataFrame(X, columns=df.drop(columns=['Sale Price']).columns)
```

```python
import statsmodels.api as sm

X_with_const = sm.add_constant(X)

model = sm.OLS(y, X_with_const)
results = model.fit()

results.summary()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap( X.corr(), vmin=-1, vmax=1, cmap=plt.cm.RdBu )
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X_with_const.values, i) for i in range(X_with_const.shape[1])]
vif["Feature"] = X_with_const.columns

vif.sort_values(by = 'VIF Factor', ascending = False)
```

```python
pv_max = 0.01

while results.pvalues.max() > pv_max:
    max_p_value_index = results.pvalues.idxmax()
    
    X = X.drop(columns=[max_p_value_index])
    
    X_with_const = sm.add_constant(X)
    model = sm.OLS(y, X_with_const)
    results = model.fit()
# end

results.summary()
```

```python
plt.figure(figsize=(15,12))
sns.heatmap( X.corr(), annot=True, vmin=-1, vmax=1, cmap=plt.cm.RdBu )
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X_with_const.values, i) for i in range(X_with_const.shape[1])]
vif["Feature"] = X_with_const.columns

vif.sort_values(by = 'VIF Factor', ascending = False)
```
