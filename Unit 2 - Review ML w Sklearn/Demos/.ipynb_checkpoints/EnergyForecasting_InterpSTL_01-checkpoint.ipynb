```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)
pd.set_option("display.precision", 2)

plt.style.use('dark_background')
```

```python
df = pd.read_csv('../../data/Hourly Energy Consumption/DAYTON_hourly.csv',
                 index_col=0,
                 header=0)

df.index = pd.to_datetime(df.index)

df = df.rename(columns={'DAYTON_MW':'mw'})

dates = pd.Series(pd.date_range(start=df.index.min(),
                                end=df.index.max(),
                                freq='H'),
                  name='date')

df = pd.merge(df, dates, left_index=True, right_on='date', how='left')
df = df.set_index('date').sort_index()
df = df.ffill()
```

```python
df1 = df.resample('w').sum()
df1 = df1.iloc[1:-1].copy()

df1.iloc[:52].plot()
```

```python
from statsmodels.tsa.seasonal import STL

y_STL = df1.mw.copy()

# Perform STL decomposition
stl = STL(y_STL, period=10)
result = stl.fit()

y_trend    = result.trend
y_seasonal = result.seasonal
y_resid    = result.resid
```

```python
plt_idx1 = 52*4

plt.figure(figsize=(15, 12))

plt.subplot(4, 1, 1)
plt.plot(y_STL.iloc[:plt_idx1], 'r', label='Original Data')
plt.plot(y_trend.iloc[:plt_idx1], 'g', label='Trend')
plt.ylim(df1.mw.min(),df1.mw.max())
plt.legend()
plt.title('Original Time Series Data')

plt.subplot(4, 1, 2)
plt.plot(y_trend.iloc[:plt_idx1], label='Trend')
plt.legend()
plt.title('Trend Component')

plt.subplot(4, 1, 3)
plt.plot(y_seasonal.iloc[:plt_idx1], label='Seasonal')
#plt.ylim(df1.mw.min(),df1.mw.max())
plt.legend()
plt.title('Seasonal Component')

plt.subplot(4, 1, 4)
plt.plot(y_resid.iloc[:plt_idx1], label='Residual')
#plt.ylim(df1.mw.min(),df1.mw.max())
plt.legend()
plt.title('Residual Component')

plt.tight_layout()
plt.show()
```

```python
df2 = df1.copy()

df2['mw_t']  = y_trend
df2['mw_ts'] = y_trend + y_seasonal

df2
```

```python
df2[:52].plot()
```

```python
df3 = pd.DataFrame()

col = 'mw_t'

df3['mw_raw'] = df2[col].resample('d').mean()
df3['mw_lin'] = df2[col].resample('d').mean().interpolate(method='linear')
df3['mw_cub'] = df2[col].resample('d').mean().interpolate(method='cubic')
df3['mw_ak']  = df2[col].resample('d').mean().interpolate(method='akima')
df3['mw_pc']  = df2[col].resample('d').mean().interpolate(method='pchip')


df3.head(15)
```

```python
plt.figure(figsize=(7,4))

plt_idx2 = 7*52*4

#col = 'mw_lin'
col = 'mw_cub'

sns.lineplot( x=df3.iloc[:plt_idx2].mw_raw.index, y=df3.iloc[:plt_idx2][col], color='b', linewidth=1, label=col )
sns.scatterplot( x=df3.iloc[:plt_idx2].mw_raw.index, y=df3.iloc[:plt_idx2].mw_raw, color='r', linewidth=0, s=50, label='mw_raw' )

plt.legend()
```

# make features

```python
df3['dayofweek'] = df3.index.dayofweek
df3['dayofmonth'] = df3.index.day
df3['week'] = df3.index.isocalendar().week.astype(int)
df3['month'] = df3.index.month
df3['quarter'] = df3.index.quarter
df3['year'] = df3.index.year

df3
```

```python
from numpy import pi

for i in range(1,7//2):
    df3[f'dayofweek_sin{i}'] = np.sin(i*2*pi*df3.dayofweek/7)
    df3[f'dayofweek_cos{i}'] = np.cos(i*2*pi*df3.dayofweek/7)
# end

for i in range(1,31//2):
    df3[f'dayofmonth_sin{i}'] = np.sin(i*2*pi*df3.dayofmonth/31)
    df3[f'dayofmonth_cos{i}'] = np.cos(i*2*pi*df3.dayofmonth/31)
# end

for i in range(1,53//2):
    df3[f'year_sin{i}'] = np.sin(i*2*pi*df3.year/53)
    df3[f'year_cos{i}'] = np.cos(i*2*pi*df3.year/53)
# end

for i in range(1,12//2+1):
    df3[f'month_sin{i}'] = np.sin(i*2*pi*df3.month/12)
    df3[f'month_cos{i}'] = np.cos(i*2*pi*df3.month/12)
# end

for i in range(1,4//2+1):
    df3[f'quarter_sin{i}'] = np.sin(i*2*pi*df3.quarter/4)
    df3[f'quarter_cos{i}'] = np.cos(i*2*pi*df3.quarter/4)
# end
```

```python
df3
```

```python
X = df3.iloc[:,5:]
y = df3['mw_cub']
```

```python
idx_split = 7*52*2
print(idx_split)

X_train = X.iloc[:-idx_split,:]
y_train = y[:-idx_split]

X_test  = X.iloc[-idx_split:,:]
y_test  = y[-idx_split:]

plt.figure(figsize=(7,3))
plt.plot( y_train, 'b' );
plt.plot( y_test, 'r' );
```

# model

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor

models = [KNeighborsRegressor( n_neighbors=25, weights='uniform' ),
          RandomForestRegressor(n_estimators=100, max_depth=5),
          RandomForestRegressor(n_estimators=100, max_depth=10),
          GradientBoostingRegressor(n_estimators=100, learning_rate=0.01, max_depth=2),
          GradientBoostingRegressor(n_estimators=100, learning_rate=0.01, max_depth=3),
          GradientBoostingRegressor(n_estimators=100, learning_rate=0.01, max_depth=5),
          GradientBoostingRegressor(n_estimators=100, learning_rate=0.01, max_depth=10)
         ]
```

```python
from sklearn.metrics import r2_score, mean_squared_error

for m in models:
    m.fit(X_train, y_train)
    
    y_train_pred = m.predict(X_train)
    y_test_pred  = m.predict(X_test)
    
    print(m)
    
    r2_train = r2_score(y_train, y_train_pred)
    print(f'r2_train:   {r2_train:.5}')

    r2_test = r2_score(y_test, y_test_pred)
    print(f'r2_test:    {r2_test:.5}')
    
    rmse_train = mean_squared_error(y_train, y_train_pred)**0.5
    print(f'rmse_train: {rmse_train:.5}')

    rmse_test = mean_squared_error(y_test, y_test_pred)**0.5
    print(f'rmse_test:  {rmse_test:.5}')
    
    mape_train = np.abs( (y_train - y_train_pred)/y_train ).mean()
    print(f'mape_train: {mape_train:.5}')

    mape_test = np.abs( (y_test - y_test_pred)/y_test ).mean()
    print(f'mape_test:  {mape_test:.5}')
    
    print()
# end
```

```python
y_train_pred = pd.Series( y_train_pred, index=y_train.index )
y_train_pred

plt.figure(figsize=(12,5))
plt.plot( y_train, 'r' );
plt.plot( y_train_pred, 'g' );
```

```python
y_test_pred = pd.Series( y_test_pred, index=y_test.index )
y_test_pred

plt.figure(figsize=(7,3))
plt.plot( y_test, 'r' );
plt.plot( y_test_pred, 'g' );
```
