```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)

plt.style.use('dark_background')
```

# Create data

```python
nPts = 1000

m = 0.75
b = 1

x = np.linspace(-3, 3, nPts)
y = m*x + b + np.random.normal(0, 0.3, nPts)
#y = 0.3*x**2 + np.random.normal(0, 0.3, nPts)
#y = np.sin(x) + np.random.normal(0, 0.3, nPts)

X = x[:,np.newaxis]

plt.scatter(x, y, label='Data', s=3);
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.gca().set_aspect('equal')
```

# Split data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

plt.scatter(X_train[:,0], y_train, label='Training Data', c='b', s=3);
plt.scatter(X_test[:,0], y_test, label='Testing Data', c='r', s=3);
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.gca().set_aspect('equal')
```

# Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from math import sqrt

# Train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# r^2
r2_train = r2_score(y_train, y_train_pred)
print(f'r^2 (train): {r2_train}')

r2_test = r2_score(y_test, y_test_pred)
print(f'r^2 (test): {r2_test}')
print()

# MSE
mse_train = mean_squared_error(y_train, y_train_pred)
print(f'MSE (train): {mse_train}')

mse_test = mean_squared_error(y_test, y_test_pred)
print(f'MSE (test): {mse_test}')
print()

# RMSE
rmse_train = sqrt(mse_train)
print(f'RMSE (train): {rmse_train}')

rmse_test = sqrt(mse_test)
print(f'RMSE (test): {rmse_test}')
print()

# MAPE
mape_train = np.mean(np.abs((y_train - y_train_pred) / y_train)) * 100
print(f'MAPE (train): {mape_train}%')

mape_test = np.mean(np.abs((y_test - y_test_pred) / y_test)) * 100
print(f'MAPE (test): {mape_test}%')
print()
```

```python
coef_sk = model.coef_
int_sk = model.intercept_
print(coef_sk, int_sk)
```

```python
plt.scatter(X_train[:,0], y_train, label='Training Data', c='b', s=3);
plt.plot(X_train[:,0], y_train_pred, c='g')
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.gca().set_aspect('equal')
```

```python
plt.scatter(X_test[:,0], y_test, label='Testing Data', c='r', s=3);
plt.plot(X_test[:,0], y_test_pred, c='g')
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.gca().set_aspect('equal')
```

# Plotting predictions vs. true values

```python
plt.scatter(y_train, y_train_pred, s=3);
plt.plot([-4,4],[-4,4],'m');
plt.xlabel('true values')
plt.ylabel('predictions')

plt.gca().set_aspect('equal')
```

# Perform cross validation

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits=5, shuffle=True)

model = LinearRegression()

for train_index, test_index in kf.split(X):
    #print(len(train_index),len(test_index),len(X))
    
    X_train = X[train_index,:]
    y_train = y[train_index]
    
    X_test  = X[test_index,:]
    y_test  = y[test_index]
    
    model.fit(X_train, y_train)
    y_test_pred = model.predict(X_test)

    r2 = r2_score(y_test, y_test_pred)
    print(r2)
```

# Perform cross validation in one step

```python
from sklearn.model_selection import cross_val_score, KFold

model = LinearRegression()

kf = KFold(n_splits=5, shuffle=True)

cv_scores = cross_val_score(model, X, y, cv=kf, scoring='r2')

cv_scores = pd.DataFrame(cv_scores)
cv_scores.describe().T
```
