```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)

plt.style.use('dark_background')
```

```python
df = pd.read_csv('../../../CSCI 4900/data/Real Estate/Real Estate Data.csv', index_col=0, header=0)
df = df.reset_index(drop=True)

df = df.rename(columns = {'Type':'Dwelling Type',
                          'OvQual':'Overall Qual',
                          'Nbhd':'Neighborhood',
                          'Built':'Year Built'})

df.info()
df.head()
```

```python
gb1 = df.groupby(["Neighborhood"])['Lot Frontage'].median().sort_values()

def fill_1(r):
    n = r.Neighborhood
    
    x = r['Lot Frontage']
    
    if pd.isna(x):
        return gb1[n]
    else:
        return x
    # end
# end

df['Lot Frontage'] = df.apply(fill_1, axis=1)
```

```python
gb2 = df.groupby(["Neighborhood"])['Garage Yr Built'].median().sort_values()

def fill_2(r):
    n = r.Neighborhood
    
    x = r['Garage Yr Built']
    
    if pd.isna(x):
        return gb2[n]
    else:
        return x
    # end
# end

df['Garage Yr Built'] = df.apply(fill_2, axis=1)
```

```python
gb3 = df.groupby(["Neighborhood"])['Masonry/Veneer Area'].median().sort_values()

def fill_3(r):
    n = r.Neighborhood
    
    x = r['Masonry/Veneer Area']
    
    if pd.isna(x):
        return gb2[n]
    else:
        return x
    # end
# end

df['Masonry/Veneer Area'] = df.apply(fill_3, axis=1)
```

```python
df = df.select_dtypes(exclude=object)

X = df.drop(columns=['Sale Price'])
y = df['Sale Price']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

scaler = StandardScaler()
X = scaler.fit_transform(X)

X = pd.DataFrame(X, columns=df.drop(columns=['Sale Price']).columns)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import Lasso, Ridge
from sklearn.metrics import r2_score, mean_squared_error

model = Lasso(alpha=300)
#model = Ridge(alpha=20)

model.fit(X_train, y_train)
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# r^2
r2_train = r2_score(y_train, y_train_pred)
print(f'r^2 (train): {r2_train}')

r2_test = r2_score(y_test, y_test_pred)
print(f'r^2 (test): {r2_test}')
print()

# MSE
mse_train = mean_squared_error(y_train, y_train_pred)
print(f'MSE (train): {mse_train}')

mse_test = mean_squared_error(y_test, y_test_pred)
print(f'MSE (test): {mse_test}')

coef = model.coef_
print(coef)

plt.figure(figsize=(6,3));
plt.bar(range(coef.shape[0]), coef);
plt.xlabel('Feature');
plt.ylabel('Coefficient Value');
#plt.ylim(-100,100);
```

```python
alphas = np.linspace(0,100000,1000)

r2s_train = []
r2s_test  = []
coefs    = []
non_zero = []

for a in alphas:
    model = Lasso(alpha=a)
    #model = Ridge(alpha=a)

    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    r2s_train.append( r2_score(y_train, y_train_pred) )
    r2s_test.append( r2_score(y_test, y_test_pred) )
    
    coefs.append( model.coef_ )
    
    non_zero.append( (model.coef_ != 0).sum() )
# end

r2s_train = np.array(r2s_train)
r2s_test = np.array(r2s_test)
coefs = np.array(coefs)
non_zero = np.array(non_zero)
```

```python
plt.plot( alphas, r2s_train, 'b' )
plt.plot( alphas, r2s_test, 'r' )
```

```python
for i in range(coefs.shape[1]):
    #plt.plot( alphas, coefs[:,i] )
    plt.semilogx( alphas, coefs[:,i] )
```
