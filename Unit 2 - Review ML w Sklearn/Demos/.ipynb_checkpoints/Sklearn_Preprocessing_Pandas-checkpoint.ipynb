```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)

plt.style.use('dark_background')
```

# Load data

```python
df = pd.read_csv('../../../CSCI 4900/data/Real Estate/Real Estate Data.csv', index_col=None, header=0)

# Clarify the names on some fields
df = df.rename(columns = {'Type':'Dwelling Type',
                          'OvQual':'Overall Qual',
                          'Nbhd':'Neighborhood',
                          'Built':'Year Built'})

df.info()
df.head()
```

# Apply standard scaling to continuous feature: 

Since there are many zeros in this column, it might be good to apply min/max scaling so that the zeros stay zeros

```python
df['Sale Price'].hist(bins=35)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# NOTE the double  brackets
salePrice2 = scaler.fit_transform(df[ ['Sale Price'] ])
salePrice2
```

```python
df['Sale Price (scaled)'] = salePrice2[:,0]
df.head(2)
```

# Apply min/max scaling to continuous feature: Garage Area

Since there are many zeros in this column, it might be good to apply min/max scaling so that the zeros stay zeros

```python
df['Garage Area'].hist(bins=35)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# NOTE the double  brackets
garageArea2 = scaler.fit_transform(df[ ['Garage Area'] ])
garageArea2
```

```python
df['Garage Area (scaled)'] = garageArea2[:,0]
df.head(2)
```

# Apply quantile transform to continuous continuous feature with outliers: Lot Area

There are some lots that are EXTREMELY large (outliers), so it would be good to use quantile transforming on these

```python
df['Lot Area'].hist(bins=55)
```

```python
from sklearn.preprocessing import quantile_transform

# NOTE the double  brackets
lotArea2 = quantile_transform(df[['Lot Area']], n_quantiles=4)
lotArea2
```

```python
df['Lot Area (quant)'] = lotArea2[:,0]
df.head(2)
```

```python
df.plot( x='Lot Area', y='Lot Area (quant)', kind='scatter')
```

# Central Air (label encode binary caterogical feature)

```python
df['CentralAir'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

LE = LabelEncoder()

# for some reason label_encoder can handle 1D
df['CentralAir'] = LE.fit_transform(df['CentralAir'])

df['CentralAir'].value_counts()
```

# Exterior Cond (ordinal encode ordinal categorical feature)

```python
df['Exterior Cond'].value_counts()
```

```python
LE = LabelEncoder()

LE.fit(df['Exterior Cond'])

# label encoding works alphabetically
# won't work as intended for these ordinal features
LE.classes_
```

```python
from sklearn.preprocessing import OrdinalEncoder

OE = OrdinalEncoder(categories=[ ['Poor', 'Fair', 'Average', 'Good', 'Excellent'] ])
OE.fit(df[ ['Exterior Cond'] ])

OE.categories_
```

```python
df['Exterior Cond (ord enc)'] = OE.transform(df[ ['Exterior Cond'] ])[:,0]

display(df['Exterior Cond (ord enc)'].value_counts().sort_index())

df.head()
```
