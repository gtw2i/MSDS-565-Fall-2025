```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)

plt.style.use('dark_background')
```

# Read data

```python
df = pd.read_csv('../../../CSCI 4900/data/Real Estate/Real Estate Data.csv', index_col=0, header=0)
df = df.reset_index(drop=True)

df = df.rename(columns = {'Type':'Dwelling Type',
                          'OvQual':'Overall Qual',
                          'Nbhd':'Neighborhood',
                          'Built':'Year Built'})

df.info()
df.head()
```

# check nulls

```python
df.isna().sum().sort_values(ascending=False)
```

# Central Air (label encode binary caterogical feature)

```python
df['CentralAir'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

LE = LabelEncoder()
df['CentralAir'] = LE.fit_transform(df['CentralAir'])

df['CentralAir'].value_counts()
```

# Exterior Cond (ordinal encode ordinal categorical feature)

```python
df['Exterior Cond'].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder

OE = OrdinalEncoder(categories=[ ['Poor', 'Fair', 'Average', 'Good', 'Excellent'] ])
OE.fit(df[ ['Exterior Cond'] ])

df['Exterior Cond'] = OE.transform(df[ ['Exterior Cond'] ])[:,0]
```

# Exterior Qual (ordinal encode ordinal categorical feature)

```python
df['Exterior Qual'].value_counts()
```

```python
OE = OrdinalEncoder(categories=[ ['Poor', 'Fair', 'Average', 'Good', 'Excellent'] ])
OE.fit(df[ ['Exterior Qual'] ])

df['Exterior Qual'] = OE.transform(df[ ['Exterior Qual'] ])[:,0]
```

# Neighborhoob (one hot encode categorical feature)

```python
df = df.replace('Bloomington Hts', 'Bloomington Heights')

df_nbhd = pd.get_dummies(df['Neighborhood'], prefix='nbhd', dtype=int, drop_first=True)
df_nbhd.head()
```

```python
df = pd.concat([df, df_nbhd], axis=1)
```

# check for null numeric columns

```python
df.select_dtypes(exclude=object).isna().sum()
```

### fill nans in Lot Frontage based on Neighborhood means

```python
gb1 = df.groupby(["Neighborhood"])['Lot Frontage'].median().sort_values()

def fill_1(r):
    n = r.Neighborhood
    
    x = r['Lot Frontage']
    
    if pd.isna(x):
        return gb1[n]
    else:
        return x
    # end
# end

df['Lot Frontage'] = df.apply(fill_1, axis=1)
```

### fill nans in 'Garage Yr Built based on Neighborhood means

```python
gb2 = df.groupby(["Neighborhood"])['Garage Yr Built'].median().sort_values()

def fill_2(r):
    n = r.Neighborhood
    
    x = r['Garage Yr Built']
    
    if pd.isna(x):
        return gb2[n]
    else:
        return x
    # end
# end

df['Garage Yr Built'] = df.apply(fill_2, axis=1)
```

### fill nans in Masonry/Veneer Area based on Neighborhood means

```python
gb3 = df.groupby(["Neighborhood"])['Masonry/Veneer Area'].median().sort_values()

def fill_3(r):
    n = r.Neighborhood
    
    x = r['Masonry/Veneer Area']
    
    if pd.isna(x):
        return gb2[n]
    else:
        return x
    # end
# end

df['Masonry/Veneer Area'] = df.apply(fill_3, axis=1)
```

# Perform modeling

```python
df = df.select_dtypes(exclude=object)

X = df.drop(columns=['Sale Price'])
y = df['Sale Price']
```

### statsmodels

```python
import statsmodels.api as sm

X_with_const = sm.add_constant(X)

model = sm.OLS(y, X_with_const)
results = model.fit()

results.summary()
```

### sklearn

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from math import sqrt

model = LinearRegression()
model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
# r^2
r2_train = r2_score(y_train, y_train_pred)
print(f'r^2 (train): {r2_train}')

r2_test = r2_score(y_test, y_test_pred)
print(f'r^2 (test): {r2_test}')
print()

# MSE
mse_train = mean_squared_error(y_train, y_train_pred)
print(f'MSE (train): {mse_train}')

mse_test = mean_squared_error(y_test, y_test_pred)
print(f'MSE (test): {mse_test}')
print()

# RMSE
rmse_train = sqrt(mse_train)
print(f'RMSE (train): {rmse_train}')

rmse_test = sqrt(mse_test)
print(f'RMSE (test): {rmse_test}')
print()
```

### cross validation

```python
from sklearn.model_selection import cross_val_score, KFold

model = LinearRegression()

kf = KFold(n_splits=5, shuffle=True)

cv_scores = cross_val_score(model, X, y, cv=kf, scoring='r2')
cv_scores
```

```python
cv_scores = pd.DataFrame(cv_scores)
cv_scores.describe().T
```

# single regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from math import sqrt

model = LinearRegression()

results = []

for i in range(len(df.columns)-1):
    model.fit(X_train.iloc[:,[i]], y_train)

    y_train_pred = model.predict(X_train.iloc[:,[i]])
    y_test_pred = model.predict(X_test.iloc[:,[i]])
    
    r2_train = r2_score(y_train, y_train_pred)
    r2_test = r2_score(y_test, y_test_pred)
    
    results.append({'Column Name': df.columns[i], 'r2_train': r2_train, 'r2_test': r2_test})
# end
results_df = pd.DataFrame(results)

results_df.sort_values(by='r2_test', ascending=False)
```

### perform pca, then model

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=16)
pca.fit(X_train)

X_pca = pca.transform(X)
X_pca
```

```python
from sklearn.model_selection import cross_val_score, KFold

model = LinearRegression()

kf = KFold(n_splits=5, shuffle=True)

cv_scores = cross_val_score(model, X_pca, y, cv=kf, scoring='r2')
cv_scores
```

```python
cv_scores = pd.DataFrame(cv_scores)
cv_scores.describe().T
```

```python
medians = []
for i in range(1,X_train.shape[1]):
    pca = PCA(n_components=i)
    pca.fit(X_train)
    X_pca = pca.transform(X)
    model = LinearRegression()
    kf = KFold(n_splits=5, shuffle=True)
    cv_scores = cross_val_score(model, X_pca, y, cv=kf, scoring='r2')
    cv_scores = pd.DataFrame(cv_scores)
    medians.append( cv_scores.median() )
# end

plt.plot(range(1,X_train.shape[1]), medians)
```
