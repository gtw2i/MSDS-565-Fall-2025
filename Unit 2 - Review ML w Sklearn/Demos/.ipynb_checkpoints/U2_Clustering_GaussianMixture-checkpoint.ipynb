# Make data

```python
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=3000,
                  centers=3,
                  cluster_std=0.4,
                  random_state=0)

X -= np.mean(X,axis=0)

%matplotlib inline

cmap = 'turbo'
MAX = np.max(np.abs(X))*1.2
plt.scatter(X[:,0], X[:,1], c='w', s=5);
plt.xlim(-MAX,MAX)
plt.ylim(-MAX,MAX)
plt.gca().set_aspect('equal')
```

# Gaussian Mixture Model w/ sklearn

### fit model

```python
from sklearn.mixture import GaussianMixture

n_components = 3
model = GaussianMixture(n_components=n_components)
model.fit(X)
```

### predict labels

```python
labels = model.predict(X)
labels
```

### compute probabilities for each label

```python
model.predict_proba(X)
```

### get means and covariances for each gaussian component

```python
weights = model.weights_
weights
```

```python
means = model.means_
means
```

```python
covariances = model.covariances_
covariances
```

### visualize components

```python
def gaussian_pdf(X, means, covariances):
    """
    Compute the probability density function (PDF) of a 2D Gaussian distribution.

    Parameters:
        X (ndarray): Input points of shape (N, 2).
        means (ndarray): Mean vector of shape (2,).
        covariances (ndarray): Covariance matrix of shape (2, 2).

    Returns:
        ndarray: Probability density values for each input point.
    """
    # Calculate the determinant of the covariance matrix
    cov_det = np.linalg.det(covariances)
    
    # Calculate the inverse of the covariance matrix
    cov_inv = np.linalg.inv(covariances)
    
    # Calculate the difference between X and the mean
    diff = X - means
    
    # Compute the exponent term in the Gaussian PDF
    exponent = -0.5 * np.sum(np.dot(diff, cov_inv) * diff)
    
    # Compute the normalization factor
    normalization = 1 / (2 * np.pi * np.sqrt(cov_det))
    
    # Compute the PDF values
    pdf_values = normalization * np.exp(exponent)
    
    return pdf_values
# end
```

```python
n_pts = 100

x = np.linspace(-MAX,MAX,n_pts)
y = np.linspace(-MAX,MAX,n_pts)

comps = np.zeros((n_pts,n_pts,n_components))
for i in range(n_components):
    
    for j in range(n_pts):
        for k in range(n_pts):
            comps[k,j,i] = weights[i]*gaussian_pdf(np.array([x[j], y[k]]),
                                                   means[i],
                                                   covariances[i])
        # end
    # end
# end

x, y = np.meshgrid(x,y)

%matplotlib qt

# Create 3D plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot 3D surface
ax.plot_surface(x, y, comps[:,:,0], color='r', alpha=0.5)
ax.plot_surface(x, y, comps[:,:,1], color='g', alpha=0.5)
ax.plot_surface(x, y, comps[:,:,2], color='b', alpha=0.5)
#ax.plot_surface(x, y, comps[:,:,3], color='m', alpha=0.5)
```

### visualize decision boundaries

```python
nGrid = 200
a1 = np.linspace(-MAX,MAX,nGrid)
a2 = np.linspace(-MAX,MAX,nGrid)

A1, A2 = np.meshgrid(a1,a2)
A1 = A1.flatten()
A2 = A2.flatten()
A = np.vstack((A1,A2)).T

B = model.predict(A)
B = B.reshape(nGrid,nGrid)
B = np.flipud(B)

extent = [-MAX,MAX,-MAX,MAX]

%matplotlib inline

plt.figure(figsize=(7,7))
plt.imshow(B, interpolation='none', extent=extent,cmap=cmap, alpha=0.4)
plt.scatter(X[:,0], X[:,1], c=labels, s=30, cmap=cmap);
plt.scatter(means[:, 0], means[:, 1], marker='^', s=50, c='w')
plt.xlim(-MAX,MAX)
plt.ylim(-MAX,MAX)
plt.gca().set_aspect('equal')
plt.show()
```
