```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)

plt.style.use('dark_background')
```

```python
n_pts = 100

x1 = np.random.rand(n_pts//2,1)
x2 = np.random.rand(n_pts//2,1) + 0.9
X = np.concatenate((x1,x2),axis=0)

y1 = np.zeros(n_pts//2)
y2 = np.ones(n_pts//2)
y = np.concatenate((y1,y2),axis=0)

plt.scatter(X[:,0],y,c=y, cmap='RdBu',s=10)
#plt.xlim(0,2)
plt.ylim(-1,2)
#plt.gca().set_aspect('equal')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

model = LogisticRegression()
model.fit(X, y)

y_pred = model.predict(X)

print(classification_report(y, y_pred))

conf_matrix = confusion_matrix(y, y_pred)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(3,2.5));
sns.heatmap(conf_matrix, annot=True, vmin=0, fmt='d', ax=ax);
```

```python
X_plot = np.linspace(X.min(),X.max(),101).reshape(-1,1)
prob_plot = model.predict_proba(X_plot)[:,1]

plt.scatter(X[:,0],y,c=y, cmap='RdBu',s=40)
plt.plot(X_plot[:,0],prob_plot)
#plt.xlim(0,2)
plt.ylim(-1,2)
#plt.gca().set_aspect('equal')
```
