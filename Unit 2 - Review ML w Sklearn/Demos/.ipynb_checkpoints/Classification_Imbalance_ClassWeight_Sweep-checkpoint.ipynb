```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)

plt.style.use('dark_background')
```

# Function for creating imbalanced data

```python
def GenerateData(n_pts1,n_pts2,center_sep,std):
    
    cov = std**2*np.eye(2)
    
    x1 = np.random.multivariate_normal(mean=[-center_sep/2,0],
                                       cov=cov,
                                       size=n_pts1)
    y1 = np.zeros(n_pts1)
    
    x2 = np.random.multivariate_normal(mean=[center_sep/2,0],
                                       cov=cov,
                                       size=n_pts2)
    y2 = np.ones(n_pts2)
    
    X = np.concatenate((x1,x2),axis=0)
    y = np.concatenate((y1,y2),axis=0)
    
    return X, y
# end
```

# Create data

```python
#np.random.seed(0)

n_pts0     = 1000
n_pts1     = 100
std        = 0.8
center_sep = 1.5

n_pts = n_pts0 + n_pts1

# generate data
X, y = GenerateData(n_pts0,
                    n_pts1,
                    center_sep,
                    std)

plt_scale = 3
plt.scatter(X[:,0], X[:,1], c=y, s=20, cmap='RdBu')
plt.xlim(-plt_scale,plt_scale)
plt.ylim(-plt_scale,plt_scale)
plt.gca().set_aspect('equal')
```

# Train/test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    stratify=y)
```

# Class weight

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import confusion_matrix, classification_report

#class_weight = None
#class_weight = 'balanced'
class_weight = { 0:1, 1:2 }

model = LogisticRegression(multi_class='ovr', class_weight=class_weight)
#model = DecisionTreeClassifier(max_leaf_nodes=10, class_weight=class_weight)
#model = RandomForestClassifier(n_estimators=100, max_leaf_nodes=10, class_weight=class_weight)

model.fit(X_train, y_train)

y_pred = model.predict(X)
y_train_pred = model.predict(X_train)
y_test_pred  = model.predict(X_test)

train_conf_matrix = confusion_matrix(y_train, y_train_pred)
test_conf_matrix  = confusion_matrix(y_test, y_test_pred)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(6,3));

sns.heatmap(train_conf_matrix, annot=True, ax=axes[0], vmin=0, fmt='d');
axes[0].set_title('Training');
sns.heatmap(test_conf_matrix, annot=True, ax=axes[1], vmin=0, fmt='d');
axes[1].set_title('Testing');
fig.suptitle(str(model))
plt.tight_layout()
plt.show()

print("Training")
print(classification_report(y_train, y_train_pred))

print("Testing")
print(classification_report(y_test, y_test_pred))
```

# Plot metric vs. class weight

```python
from sklearn.metrics import precision_recall_fscore_support

n_plot = 9

w = np.linspace(0,1,n_plot+2)[1:-1]

pre = np.zeros((n_plot,4))
rec = np.zeros((n_plot,4))
f1  = np.zeros((n_plot,4))

for i in range(n_plot):
    
    class_weight = { 0:1-w[i], 1:w[i] }
    
    model = LogisticRegression(multi_class='ovr', class_weight=class_weight)
    #model = DecisionTreeClassifier(max_leaf_nodes=10, class_weight=class_weight)
    #model = RandomForestClassifier(n_estimators=50, max_leaf_nodes=10, class_weight=class_weight)
    
    model.fit(X_train, y_train)

    y_pred = model.predict(X)
    y_train_pred = model.predict(X_train)
    y_test_pred  = model.predict(X_test)
    
    train_conf_matrix = confusion_matrix(y_train, y_train_pred)
    test_conf_matrix  = confusion_matrix(y_test, y_test_pred)

    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(6,3));

    sns.heatmap(train_conf_matrix, annot=True, ax=axes[0], vmin=0, fmt='d');
    axes[0].set_title('Training');
    sns.heatmap(test_conf_matrix, annot=True, ax=axes[1], vmin=0, fmt='d');
    axes[1].set_title('Testing');
    fig.suptitle(str(model))
    plt.tight_layout()
    plt.show()

    p, r, f, s = precision_recall_fscore_support(y_train, y_train_pred)
    pp, rr, ff, ss = precision_recall_fscore_support(y_test,  y_test_pred)
    
    pre[i,0] = p[0]
    pre[i,1] = p[1]
    pre[i,2] = pp[0]
    pre[i,3] = pp[1]
    
    rec[i,0] = r[0]
    rec[i,1] = r[1]
    rec[i,2] = rr[0]
    rec[i,3] = rr[1]
    
    f1[i,0] = f[0]
    f1[i,1] = f[1]
    f1[i,2] = ff[0]
    f1[i,3] = ff[1]
# end

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12,4));

axes[0].plot( w, pre[:,0], 'r', label='Train pre 0' )
axes[0].plot( w, pre[:,1], 'b', label='Train pre 1' )
axes[0].plot( w, rec[:,0], 'r--', label='Train rec 0' )
axes[0].plot( w, rec[:,1], 'b--', label='Train rec 1' )
axes[0].set_xlim(0,1)
axes[0].set_ylim(0,1)
axes[0].legend()

axes[1].plot( w, pre[:,2], 'r', label='Test pre 0' )
axes[1].plot( w, pre[:,3], 'b', label='Test pre 1' )
axes[1].plot( w, rec[:,2], 'r--', label='Test rec 0' )
axes[1].plot( w, rec[:,3], 'b--', label='Test rec 1' )
axes[1].set_xlim(0,1)
axes[1].set_ylim(0,1)
axes[1].legend()

#axes[2].plot( w, f1[:,0], 'r', label='Train f1 0' )
#axes[2].plot( w, f1[:,1], 'b', label='Train f1 1' )
axes[2].plot( w, f1[:,2], 'r--', label='Test f1 0' )
axes[2].plot( w, f1[:,3], 'b--', label='Test f1 1' )
axes[2].plot( w, (f1[:,2]*n_pts0+f1[:,3]*n_pts1)/n_pts, 'g--', label='Test f1 mean' )
axes[2].set_xlim(0,1)
axes[2].set_ylim(0,1)
axes[2].legend()
```
