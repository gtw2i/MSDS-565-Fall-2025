```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)
pd.set_option("display.precision", 4)

plt.style.use('dark_background')
```

# Entropy

```python
def entropy( P, b ):
    s = 0
    for p in P:
        if p > 0:
            s += -p*math.log(p,b)
        # end
    # end
    return s
# end
```

```python
p = np.zeros((100,2))
p[:,0] = np.linspace(0,1,100)
p[:,1] = 1 - p[:,0]

ents = [ entropy(x,2) for x in p ]

plt.plot( p[:,0], ents )
```

```python
n = 6

a = 2
x = np.exp( -np.arange(n) / a )
x = x / x.sum()

print(entropy(x, 2))

sns.barplot(x=np.arange(n), y=x);
plt.ylim(0,1);
```

```python
x = np.random.normal(0.5,0.5,50000)
xlim = [-10,10]
bins = 100

counts, _ = np.histogram(x, bins=bins, range=xlim)
counts = counts / counts.sum()
print(entropy(counts,2))

sns.histplot(x,bins=bins,binrange=xlim);
#plt.ylim(0,1000);
```

# Cross entropy

```python
def cross_entropy( P, Q, b ):
    s = 0
    for i in range(len(P)):
        if Q[i] > 0:
            s += -P[i]*math.log(Q[i],b)
        # end
    # end
    return s
# end
```

```python
n = 3

a = 0.0001
p = np.exp( -np.arange(n) / a )
p = p / p.sum()

b = .5
q = np.exp( -np.arange(n) / b )
q = q / q.sum()

print(cross_entropy(p, q, 2))

plt.plot(np.arange(n), p, 'b', marker='^', markersize=20, label='p');
plt.plot(np.arange(n), q, 'r', marker='^', markersize=10, label='q');
plt.ylim(0,1);
plt.legend()
```
