```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)

plt.style.use('dark_background')
```

```python
df = pd.read_csv('../../../CSCI 4900/data/Real Estate/Real Estate Data.csv', index_col=None, header=0)

df = df.rename(columns = {'Type':'Dwelling Type',
                          'OvQual':'Overall Qual',
                          'Nbhd':'Neighborhood',
                          'Built':'Year Built'})

df.info()
df.head()
```

```python
df['Dwelling Type'].value_counts()
```

```python
df['Dwelling Type'] = df['Dwelling Type'].replace({'1 STORY PUD':'1-STORY PUD',
                                                   '2 FAMILY CONVERSION':'2-FAMILY CONVERSION'})
df['Dwelling Type'].value_counts()
```

```python
df['CentralAir'].value_counts()
```

```python
def test(x):
    if x=='Y':
        return 1
    elif x=='N':
        return 0
    else:
        return np.nan
# end

df['CentralAir2'] = df['CentralAir'].apply(test)
df.head()
```

```python
def test(x):
    if type(x) == str:
        if 'STORY' in x:
            return float(x[0])
        # end
    # end
    
    return np.nan
# end

df['Floors'] = df['Dwelling Type'].apply(test)
df.head()
```

```python
gb = df.groupby(["Neighborhood"])['Lot Frontage'].median().sort_values()
gb
```

```python
def fill(r):
    n = r.Neighborhood
    
    L = r['Lot Frontage']
    
    if pd.isna(L):
        return gb[n]
    else:
        return L
    # end
# end

df.apply(fill, axis=1).dropna().hist(bins=35)
```
