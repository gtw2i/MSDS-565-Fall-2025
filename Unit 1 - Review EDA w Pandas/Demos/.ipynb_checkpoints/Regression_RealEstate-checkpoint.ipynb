```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)

plt.style.use('dark_background')
```

# Read data

```python
df = pd.read_csv('../../../CSCI 4900/data/Real Estate/Real Estate Data.csv', index_col=0, header=0)
df = df.reset_index(drop=True)

df = df.rename(columns = {'Type':'Dwelling Type',
                          'OvQual':'Overall Qual',
                          'Nbhd':'Neighborhood',
                          'Built':'Year Built'})

df.info()
df.head()
```

# Fill nans

### check for null numeric columns

```python
df.select_dtypes(exclude=object).isna().sum().sort_values(ascending=False)
```

### fill nans in Lot Frontage based on Neighborhood means

```python
gb1 = df.groupby(["Neighborhood"])['Lot Frontage'].median().sort_values()
gb1
```

```python
def fill_1(r):
    n = r.Neighborhood
    
    x = r['Lot Frontage']
    
    if pd.isna(x):
        return gb1[n]
    else:
        return x
    # end
# end

df['Lot Frontage'] = df.apply(fill_1, axis=1)
```

### fill nans in 'Garage Yr Built based on Neighborhood means

```python
gb2 = df.groupby(["Neighborhood"])['Garage Yr Built'].median().sort_values()
gb2
```

```python
def fill_2(r):
    n = r.Neighborhood
    
    x = r['Garage Yr Built']
    
    if pd.isna(x):
        return gb2[n]
    else:
        return x
    # end
# end

df['Garage Yr Built'] = df.apply(fill_2, axis=1)
```

### fill nans in Masonry/Veneer Area based on Neighborhood means

```python
gb3 = df.groupby(["Neighborhood"])['Masonry/Veneer Area'].median().sort_values()
gb3
```

```python
def fill_3(r):
    n = r.Neighborhood
    
    x = r['Masonry/Veneer Area']
    
    if pd.isna(x):
        return gb2[n]
    else:
        return x
    # end
# end

df['Masonry/Veneer Area'] = df.apply(fill_3, axis=1)
```

```python
df.select_dtypes(exclude=object).isna().sum().sum()
```

# Perform modeling

```python
df = df.select_dtypes(exclude=object)

X = df.drop(columns=['Sale Price'])
y = df['Sale Price']
```

### statsmodels multiple regression

```python
import statsmodels.api as sm

X_with_const = sm.add_constant(X)

model = sm.OLS(y, X_with_const)
results = model.fit()

results.summary()
```

# Try scaling values before linear model

### Standard scaler (won't make a difference)

```python
from sklearn.preprocessing import StandardScaler

# create scaler object
scaler = StandardScaler()

# fit scaler to our dataset (find mean, std
scaler.fit(X)

# apply z-score transformation
X1 = scaler.transform(X)

X1 = pd.DataFrame(X1, columns=X.columns)
```

```python
X1_with_const = sm.add_constant(X1)

model = sm.OLS(y, X1_with_const)
results = model.fit()

results.summary()
```

### Quantile transform (don't do it to everything)

```python
from sklearn.preprocessing import quantile_transform

X4 = quantile_transform(X, n_quantiles=100)
X4 = pd.DataFrame(X4, columns=X.columns)
```

```python
X4_with_const = sm.add_constant(X4)

model = sm.OLS(y, X4_with_const)
results = model.fit()

results.summary()
```

### sklearn

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from math import sqrt

model = LinearRegression()
model.fit(X_train, y_train)

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
# r^2
r2_train = r2_score(y_train, y_train_pred)
print(f'r^2 (train): {r2_train}')

r2_test = r2_score(y_test, y_test_pred)
print(f'r^2 (test): {r2_test}')
print()

# MSE
mse_train = mean_squared_error(y_train, y_train_pred)
print(f'MSE (train): {mse_train}')

mse_test = mean_squared_error(y_test, y_test_pred)
print(f'MSE (test): {mse_test}')
print()

# RMSE
rmse_train = sqrt(mse_train)
print(f'RMSE (train): {rmse_train}')

rmse_test = sqrt(mse_test)
print(f'RMSE (test): {rmse_test}')
print()
```

# Perform cross validation with original data

```python
from sklearn.model_selection import cross_val_score, KFold

model = LinearRegression()

kf = KFold(n_splits=5, shuffle=True)

cv_scores = cross_val_score(model, X, y, cv=kf, scoring='r2')

cv_scores = pd.Series(cv_scores)
cv_scores.sort_values()
```
