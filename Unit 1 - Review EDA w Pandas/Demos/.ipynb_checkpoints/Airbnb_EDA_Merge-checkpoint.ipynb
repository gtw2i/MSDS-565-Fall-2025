```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)

plt.style.use('dark_background')
```

# listings.csv

```python
df_list = pd.read_csv('../../data/Airbnb/listings.csv', index_col=None, header=0)

df_list.info()
df_list.head()
```

### plot lat/long

```python
sns.scatterplot(data=df_list, x='longitude', y='latitude', marker='s', s=10, color='red', alpha=0.5, linewidth=0)
```

```python
from bokeh.plotting import figure, show, output_file
from bokeh.tile_providers import get_provider, Vendors

# Define function to convert latitudes and longitudes to Mercator x and y coordinates
def merc_from_arrays(lats, lons):
    r_major = 6378137.000
    x = r_major * np.radians(lons)
    scale = x/lons
    y = 180.0/np.pi * np.log(np.tan(np.pi/4.0 + lats * (np.pi/180.0)/2.0)) * scale
    return x, y
# end

lat = df_list.latitude
lon = df_list.longitude

x, y = merc_from_arrays(lat, lon)

# HTML for plot
output_file("tile.html")

# Used to create map on plot
tile_provider = get_provider(Vendors.CARTODBPOSITRON) # Get tile provider for background map

# Create a Bokeh figure
p = figure(x_axis_type="mercator", y_axis_type="mercator")

# Add the map to the figure
p.add_tile(tile_provider)

# Plot the points on the figure
p.circle(x=x, y=y, size=5)

# Load the interactive HTML page with the plot
show(p)
```

# create a hosts dataframe from df_list (contains info on each unique host)

```python
# get one row per host
# we don't want to do this with host_name since two diferent hosts might have the same name
df_hst = df_list.drop_duplicates(subset='host_id').copy()
print(df_list.shape)
print(df_hst.shape)
```

```python
# get columns names, notice some have host_ prepended
cols = df_hst.columns

cols.sort_values()
```

```python
# get only the columns that hav ehost_ prepended
cols = cols[cols.str.startswith('host')]

# filter df to have only those columns
df_hst = df_hst[cols]
df_hst
```

## convert and clean column types

### host_since

```python
df_hst.host_since = pd.to_datetime(df_hst.host_since)
display(df_hst.host_since)
sns.displot(df_hst.host_since)
```

### host_is_superhost / host_has_profile_pic / host_identity_verified

```python
df_hst.host_is_superhost.value_counts()
```

```python
df_hst.host_is_superhost = df_hst.host_is_superhost.replace({'t':1,
                                                             'f':0})

df_hst.host_has_profile_pic = df_hst.host_has_profile_pic.replace({'t':1,
                                                                   'f':0})

df_hst.host_identity_verified = df_hst.host_identity_verified.replace({'t':1,
                                                                       'f':0})
```

### response/acceptance rate

```python
df_hst.host_response_rate.unique()
```

```python
# user-defined function
def fix_perc(x):
    if type(x) == str:
        return x[:-1]
    else:
        return np.nan
    # end
# end

df_hst.host_response_rate = df_hst.host_response_rate.apply(fix_perc).astype(float)/100
df_hst.host_response_rate
```

```python
# anonymous function
df_hst.host_acceptance_rate = df_hst.host_acceptance_rate.apply( lambda x: x[:-1] if type(x) == str else np.nan ).astype(float)/100
df_hst.host_acceptance_rate
```

### count number of listings for each host 

```python
listings = df_list.host_id.value_counts().reset_index()
listings = listings.rename(columns={'count':'listings'})
listings
```

```python
df_hst = df_hst.merge( listings, on='host_id' )
df_hst.head()
```

## count total reviews per month for each host

```python
rev_per_month = df_list.groupby(["host_id"])['reviews_per_month'].sum().sort_values(ascending=False).reset_index()
rev_per_month
```

```python
df_hst = df_hst.merge( rev_per_month, on='host_id' )
df_hst.head()
```

```python
sns.scatterplot( data=df_hst, x='listings', y='reviews_per_month' )
```

# Avg price for each host

```python
def fix(x):
    return x[1:].replace(',','')

df_list.price = df_list.price.apply(fix).astype(float)
df_list.price

#df_list.price = df_list.price.apply( lambda x: x[1:].replace(',','') ).astype(float)
```

```python
avg_price = round( df_list.groupby(["host_id"])['price'].mean().sort_values(ascending=False).reset_index(), 2 )

avg_price = avg_price.rename(columns={'price':'avg_price'})

avg_price
```

```python
df_hst = df_hst.merge( avg_price, on='host_id' )
df_hst.head()
```

## host_about length

```python
df_hst['about_len'] = df_hst['host_about'].apply( lambda x: len(x) if type(x) == str else 0 )
df_hst['about_len'].hist()
```

# calendar.csv

```python
df_cal = pd.read_csv('../../data/Airbnb/calendar.csv', index_col=None, header=0)

df_cal.info()
df_cal.head()
```

## fix types

```python
df_cal.date = pd.to_datetime(df_cal.date)
df_cal.available = df_cal.available.replace({'f':0, 't':1})
```

## plot availablity of a random listing

```python
ids = df_cal.listing_id.unique()

id1 = np.random.choice( ids, 1 )[0]

df_id1 = df_cal[ df_cal.listing_id == id1 ]

df_id1.plot( x='date', y='available' )
```

## plot avg availablity of all listings

```python
df_cal.groupby(['listing_id','date'])['available'].mean()
```

```python
pivot = df_cal.pivot_table(index='listing_id', columns='date', values='available', fill_value=0)
pivot
```

```python
# percentage of listings available on the date
pivot.mean(axis=0).plot()
```

```python
# fraction of days each listing is available
pivot.mean(axis=1).hist(bins=50)
pivot.mean(axis=1)
```
