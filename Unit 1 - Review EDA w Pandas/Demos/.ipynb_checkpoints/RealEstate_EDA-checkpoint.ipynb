```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)

plt.style.use('dark_background')
```

```python
df = pd.read_csv('../data/Real Estate/Real Estate Data.csv', index_col=None, header=0)

df.info()
df.head()
```

```python
# Clarify the names on some fields
df = df.rename(columns = {'Type':'Dwelling Type',
                          'OvQual':'Overall Qual',
                          'Nbhd':'Neighborhood',
                          'Built':'Year Built'})
```

```python
df.isna().sum().plot(kind='bar', figsize=(20,5))
```

```python
num_cols = df.select_dtypes(exclude=object).columns

for c in num_cols:
    ax = df[c].hist(bins=25, figsize=(8,4))
    ax.set_title(c)
    ax.grid(False)
    plt.show()
    print( 'nulls: ', df[c].isna().sum() )
# end
```

```python
df['Pool'] = df['Pool Area'].apply(lambda x: 1 if x > 0 else 0)

df = df.drop(['Id', 'Garage Yr Built', 'Garage Area', 'Pool Area'], axis = 1)
```

```python
obj_cols = df.select_dtypes(include=object).columns

for c in obj_cols:
    vc = df[c].value_counts()
    nulls = pd.Series( [df[c].isna().sum()], index=['nulls'] )
    vc = pd.concat( [vc,nulls] )
    
    print(c)
    print(vc)
    print()
# end
```

```python
ct = pd.crosstab(df['Dwelling Type'], df['House Style'])
display(ct)

sns.heatmap(ct)
```
